## **What is a Database?**

**Database Analogy:**

Imagine you have a large collection of books, and you want a systematic way to organize them so you can easily find, add, remove, or update details about any book. A database is a lot like a digital version of that system, but for data instead of just books.

**Definition of Database**

* A database is a structured collection of data organized and stored systematically, typically in a digital format.

* Databases are designed to store, retrieve, add, delete, and update data in an efficient and secure manner.

* They can store and manage large amounts of information, such as text, numbers, images, and multimedia files.

* They can be anything from a simple list of names saved in a file on your computer, to a massive collection of information stored on powerful servers

* Databases are typically organized around **data models**.



## **Data Models**
* Data models define the structure of the data and the relationships between different pieces of data. This structure makes it easy to query the data and find the information you need.
* Data model in databases is like a blueprint for how data is organized and how relationships between data are handled.

Data model helps you figure out:

* What kind of data you'll store.
* How you'll organize that data into tables.
* What attributes (columns) those tables will have.
* And how tables relate to one another.





## **Why are Databases Important?**
* **Efficiency and Speed**: Saving or retrieving or searching information is fast.
* **Organization**: Data is structured in tables (similar to spreadsheet rows and columns) - easy to work with.
* **Reliability** - Data recovery mechanisms ensure no data lost
* **Security** - Data access, encryption, backup ensure data protection


## **Types of Databases**

Different types of Databases exist, including: Relational, NoSQL, Graph databases, etc

We're going to focus on Relational Databases in this session.




# 1. **Relational Databases**

 * A relational database is a collection of one or more tables.

* These tables store data in rows and columns.

* A database table contains one or more rows i.e it is a collection of rows.

* A table row contains one or more columns.

* Rows represent records (e.g entry of a specific person)

* Columns represent `fields` or `attributes` of those records (e.g name, age, address etc.).

>Hence A relational database is a system that manages these tables and the relationships between them.

* Data stored in Databases can be queried using **SQL**.

## What is Structured Query Language (SQL)?
SQL is a language to communicate with databases: to ask questions, to give commands, and to organize data in meaningful ways.

> SQL allows you to "talk" to the database.
>e.g `"show me everyone aged 25" or "count how many people live in New York"`, and the database responds with the appropriate data.

### **Basic Operations:**

SQL can do a variety of tasks, including:

* **SELECT:** Retrieve data from one or more tables.
* **INSERT:** Add new records to a table.
* **UPDATE:** Change existing records in a table.
* **DELETE:** Remove records from a table.
 - and many more specialized operations.

#### **Step 5: Insert New Record via SQL**

* Insert statement inserts a row into a table
```
INSERT INTO Users(name, email) VALUES ('Fatma Smarty', 'fatma.smarty@email.com')
```

>`INSERT INTO Users(name, email)`: specifies you want to insert data into the `Users` table and particularly into the `name` and `email` columns.

>`VALUES ('Fatma Smarty', 'fatma.smarty@email.com')`: Defines values you want to insert. Each value corresponds to the column mentioned in the INSERT INTO part.

#### **Step 6: SQL Delete**
* Deletes a row in a table based on a selection criteria

```
DELETE FROM Users WHERE name = 'John Doe';

```

**Explanation:**

> `DELETE FROM Users`: This specifies that you want to delete data from the `Users` table.

> `WHERE name = 'John Doe'`: This is a condition that selects which record(s) to delete. In this case, it's looking for records where the name column has the value "John Doe". Only those records that meet this condition will be deleted.

image.png

#### **Step 7: SQL Update**

* Allows the updating of a field with a where clause

```
UPDATE Users
SET email = 'john.doe_updated@email.com'
WHERE name = 'John Doe';

```

**Explanation:**

> `UPDATE Users`: Specifies the table (Users) you want to update.

> `SET email` = 'john.doe_updated@email.com': Defines what you want to update in the record. E.g sets the `email` column to "john.doe_updated@email.com".

>`WHERE name` = 'John Doe': This is a condition that determines which record(s) to update. In this case, it will look for records where the `name` column has the value "John Doe". Only those records that meet this condition will have their email updated.

image.png

#### **Step 8: Retrieve Records: Select**

* The select statement retrieves a group of records - you can either retrieve all the records or a subset of the records with a WHERE clause.

* Retrieve all emails of users named "John Doe" from the `Users` table.
```
SELECT email
FROM Users
WHERE name = 'John Doe'
```

**Explanation:**

> SELECT email: Specifies which column you want to retrieve. In this case, it's the email column. If you want to retrieve multiple columns, you can separate them with commas. If you want to retrieve all columns, you'd use an asterisk (*), i.e., SELECT *.

> FROM Users: Specifies from which table you're selecting, which is the Users table in this case.

> WHERE name = 'John Doe': This is a condition that determines which record(s) to retrieve. In this case, it will select records where the name column has the value "John Doe".


**Examples**
Retrieve all records and all columns from the Users table:
```
SELECT *
FROM Users

```

Retrieve names of users who have an email ending with "@email.com":

```
SELECT name
FROM Users
WHERE email LIKE '%@email.com'
```
>The % in the LIKE clause acts as a wildcard, matching any sequence of characters.

Retrieve both name and email for all users:
```
SELECT name, email
FROM Users;

```

## **Extra Notes**

## Database Components


### Tables
Tables are the basic building blocks of a database. They represent the structured format for organizing and storing data. Tables consist of rows (records) and columns (fields) defining the data structure and attributes of the stored entities.

## Schema

The schema defines the logical structure and organization of the data base. It specifies the tables, their relationships, and the constraints that govern the data. The schema acts as a blueprint for data storage and defines the data types, primary keys, foreign keys, and other constraints.

## Queries
Queries retrieve, manipulate, and analyze data stored in the data base. They allow users to extract specific information based on criteria, perform calculations, and aggregate data.

## Indexes
Indexes are data structures that improve the speed of data retrieval operations. They provide a quick reference to the location of data within a table, allowing for faster search and retrieval based on specific columns or keys.



## Database Management System (DBMS)

The DBMS is the software that manages the database. It provides an interface for users to interact with the database, handles data storage, retrieval, and manipulation, enforces security and access controls, and ensures data integrity and consistency

## Data Manipulation Language (DML)
DML is a language that enables users to insert, update, delete, and modify data in the database. Common DML languages include SQL (Structured Query Language).

## Data Definition Language (DDL)
DDL is a language used to define the structure and properties of the data base. It includes commands for creating tables, defining relationships, setting constraints, and managing the schema.

## Data Backup and Recovery
Data base systems often provide mechanisms for backing up data and restoring it in case of data loss or system failures. Backup strategies include full backups, incremental backups, and point-in-time recovery options.

## Security and Access Controls
Data base systems implement security measures to control access to data and ensure data privacy. User authentication, authorization, and encryption techniques are commonly employed to protect the integrity and confidentiality of the data.
