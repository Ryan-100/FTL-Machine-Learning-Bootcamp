## **What is a regular expression and why is it useful?**

* A regular expression is a sequence of characters that specifies a match pattern in text
* Regular expressions ("regex"), provide a way to search text for patterns
* In Python, the `re` module provides the capability to work with regex.

* **Efficiency and Speed:** regex operations are optimized and achieve more in a single line of code.
* **Versatility Across Platforms and Languages:**  Principles of regular expressions apply across many programming languages (Python, R, Java, C++, and more)

```python
text="abc akc adc adbc" #objective . Find a word in the text that starts with a and ends with c. And inbetween a and c you have 2 characters of any type
```

### **Matching multiple copies**

```python
text="Meier Maier Maayer Maaaaayer MYER mYER" 
pattern=r"M[ae]+[iy]er" # Searching for a word, that starts with 'M', then followed by any number of "a" either 'a' or 'e', followed
#by either 'i' or 'y' and ends with 'er'
re.findall(pattern,text)
```

```python
#search for exactly 2 numbers of a's or e's
text="Meier Maier Maayer Maaaaayer MYER mYER"

pattern=r"M[ae]{2}[iy]er" #
re.findall(pattern,text)
```

# **Exercise:** Extracting email addresses from text

Given a text below,write a regular expression pattern to extract all emails from a text.

email_text = "Contact us at support@example.com or at helpdesk@mysite.org."

**Expected Output**
['support@example.com', 'helpdesk@mysite.org']


```python
email_text = "Contact us at sup..port@example.com or at helpdesk@mysite.org."

pattern = r'[\w\.-]+@[\w]+\.[a-z]+'

print(re.findall(pattern, email_text))
```

### **How to Fetch a Web Page?**

Python provides `requests` library to make HTTP requests to fetch web pages.

**HTTP Request methods**
HTTP methods define the type of action the request wishes to perform:

Most common include:

* **POST (Create)**: It means to sending data to a server to create some new content.

* **GET (Read)**: it means to retrieve data from the server.

* **PUT (Update)**: it means to request a change of the name of the task.

* **DELETE (Delete)**: it means to request the server to delete some content


> HTTP stands for "Hypertext Transfer Protocol." which is a set of rules for how web browsers (like Chrome or Firefox) and servers (computers where websites are hosted) talk to each other.

### **2.2 Parsing XML**
XML (eXtensible Markup Language) is commonly used for storing and transporting data.

* With Python, you can parse XML documents and extract the information you need.

```python
# Sample XML data and parsing with BeautifulSoup
xml_data = """
<data>
    <country name="Liechtenstein">
        <rank>1</rank>
    </country>
    <country name="Singapore">
        <rank>4</rank>
    </country>
    <country name="Panama">
        <rank>68</rank>
    </country>
</data>
"""

soup = BeautifulSoup(xml_data, 'xml')
for country in soup.find_all('country'):
    print(f"Country: {country['name']}, Rank: {country.rank.string}")
```

 **2.4 Introducing JSON**

JSON (JavaScript Object Notation) JSON is a lightweight format for storing and transporting data.

* In Python, the built-in `json` module allows you to encode and decode JSON data easily.

```python
import json

json_data = """
{
    "users": [
        {"name": "Alice", "age": 30},
        {"name": "Bob", "age": 25}
    ]
}
"""

data = json.loads(json_data)
for user in data["users"]:
    print(f"Name: {user['name']}, Age: {user['age']}")
```

## **Accessing Data via API**

United Nations Statistics Division SDG API provides you to able to explore the official SDG data reported by the custodiam agencies

```python
# Import the necessary libraries
import requests  # For making HTTP requests.
import json      # For encoding and decoding JSON data.

# Send a GET request to fetch the data from the provided API URL.
api_url = "https://unstats.un.org/SDGAPI/swagger/v1/swagger.json"
response = requests.get(api_url)

# Encode the response text to bytes and then decode it to remove the UTF-8 Byte Order Mark (BOM) if present.
# This step ensures that the data is correctly formatted for JSON parsing.
decoded_data = response.text.encode().decode('utf-8-sig')

# Parse the decoded data string into a Python data structure (dictionary or list).
json_df = json.loads(decoded_data)
```
