### Looping through Strings

* **Using a While statement**


**Example: Counting Vowels**

```python
def count_vowels_while(input_string: str) -> int:
    vowels = "aeiou"
    count = 0
    i = 0

    # Convert the string to lowercase to make the comparison case insensitive
    input_string = input_string.lower()

    while i < len(input_string):
        if input_string[i] in vowels:
            count += 1
        i += 1
    return count


# Test the functions
word = "HELLO"

print(count_vowels_while(word))
```

* **Using a For statement**

```python
def count_vowels_for(input_string: str) -> int:
    vowels = "aeiou"
    count = 0

    # Convert the string to lowercase to make the comparison case insensitive
    input_string = input_string.lower()

    for char in input_string:
        if char in vowels:
            count += 1
    return count

word = "Abracadabra"
print(f"Number of vowels in '{word}' using for loop: {count_vowels_for(word)}")
```

```python
def is_polindrome_while(input_string: str) -> bool:
    
    input_string = ''.join(c for c in input_string if c.isalnum()).lower()
    
    left = 0
    right = len(input_string) - 1
    
    while left < right:
        if input_string[left] !=input_string[right]:
            return False
        left +=1
        right -=1
        
    return True

print(is_polindrome_while("A man, a plan, a canal, Panama" ))
```


There are several modes in which a file can be opened:

* "r" : Read mode (default). Opens the file for reading.
* "w" : Write mode. Opens the file for writing (and creates it if it doesn't exist). If the file already exists, it truncates the file.
* "a" : Append mode. Opens the file for writing but will append to the end of the file if it exists.
* "b" : Binary mode. Reads/Writes files in binary format, used especially for non-text files like images, executables, etc.

## **Writing Files**

To write to a file, you'll typically open it in **"write"** or **"append"** mode.




**Write Mode:**

* If the file already exists erase its content or replace new content with old content.

*  If the file doesn't exist, opening it in "write" mode will create a new empty file.

```python
fname = basepath + 'ex.txt'

# Open the file with write mode. This will create a new file or overwrite an existing one.
file = open(fname, "w")

file.write("We love programming\n")
file.write("We are Frontier Tech Leaders!\n")

file.close()
```

```python
fh = open(fname , 'r')

fh.read()
```

**Write to Existing File (Append Mode)**
* If the file already exists, opening it in "append" mode will not alter its contents. Append new content to end of the file, preserving existing data.

* If the file doesn't exist, opening it in "append" mode will create a new empty file.


```python
# Define the file path
file_path = basepath + "zen.txt"

# Use the 'with' statement to ensure the file gets closed properly
with open(file_path, 'a') as file_handle:
    file_handle.write("\nThis is all true!")  # Append a new line and some text
```

```python
fh = open(file_path , 'r')
fh.read()
```

```python
# Ensure the file exists and has content
with open(file_path, 'a+') as file:
    # Move the cursor to the beginning of the file and print its content
    file.seek(0)
    print("Before appending:")
    print(file.read())

# Append text to the file
with open(file_path, 'a') as file:
    file.write("\nThis is the appended text.")

# Check the content of the file after appending
with open(file_path, 'r') as file:
    print("\nAfter appending:")
    print(file.read())
```

**Closing a File**
After performing operations on a file, it's a good practice to close it using the close() method:

**Using 'with' Statements**
A more Pythonic way to work with files is to use the *with statement*. It ensures that the file is properly closed once it's no longer needed:

```python
with open(file_path, "r") as file:
    content = file.read()
    print(content)
```

**Note:** No need to explicitly call close() when using with.
